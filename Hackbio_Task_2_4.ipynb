```python
#importing
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
#opening the dataset
from posixpath import sep
sift_file_url = "https://raw.githubusercontent.com/HackBio-Internship/public_datasets/main/R/datasets/sift.tsv"
sift_df = pd.read_csv(sift_file_url, sep = "\s+", engine = "python")
sift_df
```

```python
#opening the other dataset
foldx_file_url = "https://raw.githubusercontent.com/HackBio-Internship/public_datasets/main/R/datasets/foldX.tsv"
foldx_df = pd.read_csv(foldx_file_url, sep = '\s+', engine = "python")
foldx_df
```

```python
#created a new column in sift_df
sift_df["specific_Protein_aa"] = sift_df["Protein"] + "_" + sift_df["Amino_Acid"]
sift_df
```

```python
#created a new column in foldx_df
foldx_df["specific_Protein_aa"] = foldx_df["Protein"] + "_" + foldx_df["Amino_Acid"]
foldx_df
```

```python
#Using the specific_Protein_aa column, merge sift and foldx dataset into one final dataframe.

merged_df = pd.merge(sift_df, foldx_df, on = "specific_Protein_aa")
merged_df
```

```python
#eliminate duplicates columns which have the same values, regardless of the column name.
final_df = merged_df.T.drop_duplicates().T
final_df
```

```python
#reordering columns using iloc
df_ = final_df.iloc[:, [0,1,3,2,4]]
print("Reordered Dataset:")
df_
```

```python
#Using the criteria above, Find all mutations that have a SIFT score below 0.05 and FoldX Score above 2
 #(i.e: Mutations that affect both structure and function)
#Study the amino acid substitution nomenclature

mutations_df = df_.query('sift_Score<0.05 and foldX_Score>2')
print("Dataset that shows the mutations that affect both structure and function:")
mutations_df
```

```python
#Investigate for the amino acid that has the most functional and structural impact

#Hint: Using the amino acid column, find a way to select the first amino acid.
#Extract the first amino acid from the amino acid column and store it in a new column
mutations_df['First_Amino_Acid'] = mutations_df['Amino_Acid_x'].str[0]
mutations_df

```

```python
# calculate the frequency of each first amino acid to identify the most frequent one
#generate a frequency table for all the amino acids
amino_acid_frequency = mutations_df['First_Amino_Acid'].value_counts()
amino_acid_frequency
```

```python
print(f"The amino acid that has the most functional and structural impact is {amino_acid_frequency.index[0]}, which stands for Glycine.")
```

```python
#Using the amino frequency table above,
#generate a barplot and pie chart to represent the frequency of the amino acids.

#This is the piechart

fig = px.pie(labels = amino_acid_frequency.index, values = amino_acid_frequency.values,
             title = "Amino Acids Frequency",
             names = amino_acid_frequency.index,
             hole = 0.5,
             color = amino_acid_frequency.index,

             color_discrete_sequence=px.colors.sequential.RdBu)



fig.update_traces(textposition = 'outside', textinfo = 'percent + label')

fig.update_layout(
    autosize=False,
    width=700,
    height=700
)



fig.show()
```

```python
#This is the barplot

fig = px.bar(x = amino_acid_frequency.index,
        y = amino_acid_frequency.values,
        color = amino_acid_frequency.values,
        color_continuous_scale='Bluered_r',
             title = "Amino Acids Frequency",
             labels = {
                 "x": "Amino Acid",
                 "y": "Frequency"
             }
             )

fig.show()
```

**About high occurence amino acids**

High occurrence amino acids suggest that the amino acid is at a critical position and that mutations at these positions more likely affect structure and functions. Also, looking at the specific amino acids, we can find mostly non-polar once like alanine, leucine, valine and phenylalanine, which are hydrophobic and form interactions within the protein's hydrophobic core, thus stabilising the structure. Also, alanine and leucine are known to form alpha helices.
